```python
import hashlib

f= open("hash1.txt","w")

m = hashlib.sha256()
m.update('The quick brown fox jumps over the lazy dog'.encode('utf-8'))
f.write(m.hexdigest())

f.close()
```

```python
import hashlib

f= open("hash2.txt","w")

m = hashlib.sha256()
m.update('The quick brown fox jumps over the lazy cog'.encode('utf-8'))
f.write(m.hexdigest())

f.close()
```

```python
with open('hash1.txt', 'r') as f1, open('hash2.txt', 'r') as f2:
        hash1 = f1.read()
        hash2 = f2.read()

if hash1 == hash2:
    print("해시값이 동일합니다.")
else:
    print("해시값이 다릅니다.")
```
